```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def conv(value):
    if value == 1:
        return  "yes"
    else:
        return  "no"
    end
        
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customer by churn")
plt.show()
```

```python
# Compute counts
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# Compute percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Create figure
plt.figure(figsize=(4, 4))

# Plot stacked bars
bars_no = plt.bar(ct.index, ct_percent["No"], label="No Churn", color="#4CAF50")
bars_yes = plt.bar(ct.index, ct_percent["Yes"], bottom=ct_percent["No"], label="Churn", color="#F44336")

# Add percentage labels
for idx in ct.index:
    # Label for No Churn
    plt.text(
        idx,
        ct_percent["No"][idx] / 2,
        f"{ct_percent['No'][idx]:.1f}%",
        ha="center",
        va="center",
        color="white",
        fontsize=11,
        fontweight="bold"
    )
    # Label for Churn
    plt.text(
        idx,
        ct_percent["No"][idx] + ct_percent["Yes"][idx] / 2,
        f"{ct_percent['Yes'][idx]:.1f}%",
        ha="center",
        va="center",
        color="white",
        fontsize=11,
        fontweight="bold"
    )

# Custom x-axis labels
plt.xticks(ct.index, ["Not Senior Citizen", "Senior Citizen"])

# Axis labels and title
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage of Customers")
plt.title("Churn Percentage by Senior Citizen Status")
plt.ylim(0, 100)
plt.legend(title="Churn Status", bbox_to_anchor = (0.9,0.9))

plt.tight_layout()
plt.show()
```

```python
# Columns you want to plot
cols = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'StreamingMovies'
]

# Set the figure size
fig, axes = plt.subplots(
    nrows=5, ncols=3,   # 5 rows x 3 columns grid (15 slots)
    figsize=(18, 20)
)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop over columns and axes
for i, col in enumerate(cols):
    sns.countplot(
        x=col,
        data=df,
        ax=axes[i],
        order=df[col].value_counts().index , hue = "Churn" # Order by frequency (optional)
    )
    axes[i].set_title(f'Count of {col}', fontsize=14)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)

# If fewer plots than subplots slots, remove extra axes
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#These countplots show how customer churn relates to different attributes like gender, senior citizenship, and internet services. In most categories, the number of customers who did not churn is higher than those who did. Notably, factors like Fiber optic internet and no online security seem associated with higher churn counts. The visualization helps quickly compare churn patterns across key customer characteristics.
